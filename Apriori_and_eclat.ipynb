```python
transaction = []
for i in range(dataset_original.shape[0]):
    transaction.append([str(dataset_original.values[i,j]) for j in range(dataset_original.shape[1])])
transaction = np.array(transaction)
```

```python
transaction
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction).transform(transaction)
dataset_f = pd.DataFrame(te_ary, columns=te.columns_)
dataset_f
```

```python
dataset_f
```

```python
first30 = df_table["items"].head(120).values
dataset = dataset_f.loc[:,first30]
dataset.shape
```

```python
dataset
```

```python
from mlxtend.frequent_patterns import apriori

res=apriori(dataset,min_support=0.05, use_colnames=True,verbose=3)
res
```

```python
from mlxtend.frequent_patterns import association_rules
rul=association_rules(res, metric="lift", min_threshold=1)
rul
```

```python
rul2=association_rules(res, metric="confidence", min_threshold=0.2)
rul2
```

```python
rul.sort_values("lift",ascending=False)
```

## ECLAT

```python
from pyECLAT import Example2
```

```python
dataset = Example2().get()
dataset
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
eclat = ECLAT(data=dataset)
eclat.df_bin
```

```python
items_total = eclat.df_bin.astype(int).sum(axis=0)
items_total
```

```python
items_per_transaction = eclat.df_bin.astype(int).sum(axis=1)
items_per_transaction
```

```python
df = pd.DataFrame({'items': items_total.index, 'transactions': items_total.values})
df_table = df.sort_values("transactions", ascending=False)

#  Top 5 most popular products/items
df_table.head(20).style.background_gradient(cmap='Blues')
```

```python
import plotly.express as px
# to have a same origin
df_table["all"] = "Tree Map"
# creating tree map using plotly
fig = px.treemap(df_table, path=['all', "items"], values='transactions',
                  color=df_table["transactions"], hover_data=['items'],
                  color_continuous_scale='Blues',
                )
# ploting the treemap
fig.show()
```

```python
min_support = 5/100
min_combination = 2

max_combination = max(items_per_transaction)

rule_indices, rule_supports = eclat.fit(min_support=min_support,
                                                 min_combination=min_combination,
                                                 max_combination=max_combination,
                                                 separator=' & ',
                                                 verbose=True)
```

```python
result = pd.DataFrame(rule_supports.items(),columns=['Item', 'Support'])
result.sort_values(by=['Support'], ascending=False)
```
